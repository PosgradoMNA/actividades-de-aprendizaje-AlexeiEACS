### Ejercicio 7

```python
from sklearn.decomposition import PCA

pca_model = PCA()

pca_model.fit(df)
```

```python
varianza_explicada = pca_model.explained_variance_ratio_
pd.DataFrame(
    data   = {'var_individual':varianza_explicada, 'var_acumulada':np.cumsum(varianza_explicada)},
    index  = [f'PC_{x+1}' for x in range(len(varianza_explicada))]
)
```

Se puede ver que X12 es la que por si sola podría ser un componente ya que ahí el vector hace la explicación de casi el 99% de la varianza total.

```python
pc_exp = pd.DataFrame(
    data    = pca_model.components_,
    columns = df.columns,
    index  = [f'PC_{x+1}' for x in range(len(varianza_explicada))]
)
pc_exp
```

### Ejercicio 8

```python
onepc = PCA(n_components=2)

new_df = onepc.fit_transform(df)
new_df = pd.DataFrame(
    new_df,
    columns = ['PC1', 'PC2'],
    index   = df.index
)
new_df.hist()
```

Con el heatmap en las componentes podemos ver que entre más cercano es a 1 menos variación tiene y está casi centrado en 0 con valor de 1 que es que tanta importancia tuvo.
Como podemos ver hacía el centro cada componente se empieza a abrir más representando así como es que toma menore relevancia con valores cada vez más dispersos

```python
sns.heatmap(pc_exp, cmap='BuPu')
```

Se nota lo que ya veíamos numericamente que una sola componente es necesaria pra la explicación de la varianza.

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(6, 4))
ax.bar(
    x      = np.arange(pca_model.n_components_) + 1,
    height = pca_model.explained_variance_ratio_
)
for x, y in zip(np.arange(len(df.columns)) + 1, pca_model.explained_variance_ratio_):
    label = round(y, 2)
    ax.annotate(
        label,
        (x,y),
        textcoords="offset points",
        xytext=(0,10),
        ha='center'
    )

ax.set_xticks(np.arange(pca_model.n_components_) + 1)
ax.set_ylim(0, 1.1)
ax.set_title('Porcentaje de varianza explicada por cada componente')
ax.set_xlabel('Componente principal')
ax.set_ylabel('Por. varianza explicada');
```

Con la de codo vemos que también es claro que solo es necesario una

```python
prop_varianza_acum = pca_model.explained_variance_ratio_.cumsum()

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(6, 4))
ax.plot(
    np.arange(len(df.columns)) + 1,
    prop_varianza_acum,
    marker = 'o'
)

for x, y in zip(np.arange(len(df.columns)) + 1, prop_varianza_acum):
    label = round(y, 2)
    ax.annotate(
        label,
        (x,y),
        textcoords="offset points",
        xytext=(0,10),
        ha='center'
    )
    
ax.set_ylim(0, 1.1)
ax.set_xticks(np.arange(pca_model.n_components_) + 1)
ax.set_title('Porcentaje de varianza explicada acumulada')
ax.set_xlabel('Componente principal')
ax.set_ylabel('Por. varianza acumulada');
```
